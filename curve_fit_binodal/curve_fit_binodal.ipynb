```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [10,6]
plt.rcParams['figure.dpi'] = 100

import scipy.optimize as opt
import pandas as pd
```

```python
data1 = pd.read_csv('Sample_1 100_30 01.csv',skiprows = 6)
list1 = data1.values.tolist()
data2 = pd.read_csv('Sample_2 100_30 01.csv',skiprows = 6)
list2 = data2.values.tolist()
data3 = pd.read_csv('Sample_3 100_30 01.csv',skiprows = 6)
list3 = data3.values.tolist()
data4 = pd.read_csv('Sample_4 100_30 01.csv',skiprows = 6)
list4 = data4.values.tolist()

wavelength1 = []
intensity1 = []
wavelength2 = []
intensity2 = []
wavelength3 = []
intensity3 = []
wavelength4 = []
intensity4 = []

for i in list1:
    wavelength1.append(i[0])
    intensity1.append(i[1])
for i in list2:
    wavelength2.append(i[0])
    intensity2.append(i[1])
for i in list3:
    wavelength3.append(i[0])
    intensity3.append(i[1])
for i in list4:
    wavelength4.append(i[0])
    intensity4.append(i[1])
```

```python
w_fit1 = wavelength1[135:161]
i_fit1 = intensity1[135:161]

w_fit2 = wavelength2[140:170]
i_fit2 = intensity2[140:170]

w_fit3 = wavelength3[140:176]
i_fit3 = intensity3[140:176]

w_fit4 = wavelength4[140:176]
i_fit4 = intensity4[140:176]
```

```python
def f(x,a,b,c):
    return a*((x + b)**2) + c

a0 = -0.5
b0 = -150
c0 = 3000

fit1 = opt.curve_fit(f,w_fit1,i_fit1,p0 = [a0,b0,c0])
fit2 = opt.curve_fit(f,w_fit2,i_fit2,p0 = [a0,b0,c0])
fit3 = opt.curve_fit(f,w_fit3,i_fit3,p0 = [a0,b0,c0])
fit4 = opt.curve_fit(f,w_fit4,i_fit4,p0 = [a0,b0,c0])

a1 = fit1[0][0]
b1 = fit1[0][1]
c1 = fit1[0][2]
a2 = fit2[0][0]
b2 = fit2[0][1]
c2 = fit2[0][2]
a3 = fit3[0][0]
b3 = fit3[0][1]
c3 = fit3[0][2]
a4 = fit4[0][0]
b4 = fit4[0][1]
c4 = fit4[0][2]

f1 = f(w_fit1,a1,b1,c1)
f2 = f(w_fit2,a2,b2,c2)
f3 = f(w_fit3,a3,b3,c3)
f4 = f(w_fit4,a4,b4,c4)
plt.plot(wavelength1,intensity1,marker = '',linestyle = '-',label = 'Spectra of first binodal')
plt.plot(w_fit1,f1,label = 'Fitted curve for first binodal')
plt.plot(wavelength3,intensity3,marker = '',linestyle = '-',label = 'Spectra of third binodal')
plt.plot(w_fit3,f3,label = 'Fitted curve for third binodal')

plt.legend(loc = 'best')
plt.title('Prodan Emission Spectra of binodals')
plt.xlabel('Wavelength, $\lambda$ [nm]')
plt.ylabel('Intensity')
plt.show()
```

```python
plt.plot(wavelength2,intensity2,marker = '',linestyle = '-',label = 'Spectra of second binodal')
plt.plot(w_fit2,f2,label = 'Fitted curve for second binodal')
plt.plot(wavelength4,intensity4,marker = '',linestyle = '-',label = 'Spectra of fourth binodal')
plt.plot(w_fit4,f4,label = 'Fitted curve for fourth binodal')

plt.legend(loc = 'best')
plt.title('Prodan Emission Spectra of binodals')
plt.xlabel('Wavelength, $\lambda$ [nm]')
plt.ylabel('Intensity [a.u.]')
plt.show()
```
