```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [10,6]
plt.rcParams['figure.dpi'] = 100

import scipy.optimize as opt
import pandas as pd
```

```python
data1 = pd.read_csv('3W top.csv',skiprows = 6)
list1 = data1.values.tolist()
data2 = pd.read_csv('3W btm 300_10.csv',skiprows = 6)
list2 = data2.values.tolist()


wavelength1 = []
intensity1 = []
wavelength2 = []
intensity2 = []


for i in list1:
    wavelength1.append(i[0])
    intensity1.append(i[1])
for i in list2:
    wavelength2.append(i[0])
    intensity2.append(i[1])
```

```python
w_fit1 = wavelength1[128:150]
i_fit1 = intensity1[128:150]

w_fit2 = wavelength2[144:186]
i_fit2 = intensity2[144:186]
```

```python
def f(x,a,b,c):
    return a*((x + b)**2) + c

a0 = -0.5
b0 = -150
c0 = 3000

fit1 = opt.curve_fit(f,w_fit1,i_fit1,p0 = [a0,b0,c0])
fit2 = opt.curve_fit(f,w_fit2,i_fit2,p0 = [a0,b0,c0])


a1 = fit1[0][0]
b1 = fit1[0][1]
c1 = fit1[0][2]
a2 = fit2[0][0]
b2 = fit2[0][1]
c2 = fit2[0][2]

f1 = f(w_fit1,a1,b1,c1)
f2 = f(w_fit2,a2,b2,c2)

plt.plot(wavelength1,intensity1,marker = '',linestyle = '-',label = 'Spectra of toluene rich top')
plt.plot(w_fit1,f1,label = 'Fitted curve for toluene rich top')
plt.plot(wavelength2,intensity2,marker = '',linestyle = '-',label = 'Spectra of water rich bottom')
plt.plot(w_fit2,f2,label = 'Fitted curve for water rich bottom')

plt.legend(loc = 'best')
plt.title('Prodan Emission Spectra of third tie line')
plt.xlabel('Wavelength, $\lambda$ [nm]')
plt.ylabel('Intensity')
plt.show()
```
